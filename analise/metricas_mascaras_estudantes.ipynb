```python
import json
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
import sys
sys.path.append("/home/apacheco/code/LIFE/utils")

from seg_metrics import compute_metrics

%load_ext autoreload
%autoreload 2
```

```python
PATH_MASK_PAD = "/home/apacheco/Datasets/PAD_segmentation/all-mask"
PATH_MASK_DEN = "/home/apacheco/Datasets/masks_students"

match_ids = None
with open("match_exato.json", 'r') as f:    
        match_ids = json.load(f)   

print(f"Quantidade de imagens com match: {len(match_ids)}")
```

```python
count = 1
iou_list = list()
dice_list = list()
pacc_list = list()

for id_pad, id_den in match_ids.items():
    
    print(f"\n- Avaliacao {count}:")
    print(f"- PAD: {id_pad} | DEN: {id_den}") 
    count += 1
    
    mask_pad = cv2.imread(os.path.join(PATH_MASK_PAD, id_pad.replace(".png", "_segmentation.png")), cv2.IMREAD_GRAYSCALE)  
    mask_stud = cv2.imread(os.path.join(PATH_MASK_DEN, id_den.replace(".png", "_mask.png")), cv2.IMREAD_GRAYSCALE)  
    
    if mask_pad is None:
        print("- WARN: imagem do PAD não encontrada")
        continue
    if mask_stud is None:
        print("- WARN: imagem do estudante não encontrada")
        continue
    
    print("- Metrics:")
    metrics = compute_metrics(mask_stud, mask_pad, metrics=("iou", "dice", "pixel_accuracy"), 
                              verbose=True)
    
    iou_list.append(metrics["iou"])
    dice_list.append(metrics["dice"])
    pacc_list.append(metrics["pixel_accuracy"])
```

```python
print(f"IoU: {np.mean(iou_list)} - {np.std(iou_list)}")
print(f"DICE: {np.mean(dice_list)} - {np.std(dice_list)}")
print(f"PACC: {np.mean(pacc_list)} - {np.std(pacc_list)}")
```
